## Preprocessing

```python
processed_data = raw_data.copy()
```

```python
processed_data = processed_data.drop('ID', axis=1)
```

```python
dummies = pd.get_dummies(processed_data['Reason for Absence'], drop_first=True)
processed_data["Reason_1"] = dummies.iloc[:, 0:14].max(axis=1)
processed_data["Reason_2"] = dummies.iloc[:, 14:17].max(axis=1)
processed_data["Reason_3"] = dummies.iloc[:, 17:20].max(axis=1)
processed_data["Reason_4"] = dummies.iloc[:, 20:].max(axis=1)
```

```python
processed_data = processed_data.drop('Reason for Absence', axis=1)
```

```python
processed_data['Education'] = processed_data['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
processed_data['Date'] = pd.to_datetime(processed_data['Date'], format = '%d/%m/%Y')
```

```python
processed_data['Month Value'] = processed_data['Date'].apply(lambda x: x.month)
```

```python
processed_data['Day of the Week'] = processed_data['Date'].apply(lambda x: x.weekday())
```

```python
processed_data = processed_data.drop("Date", axis=1)
```

```python
processed_data = processed_data[['Reason_1',
       'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']]
```

```python
processed_data.to_csv('Absenteeism_preprocessed.csv', index=False)
```
